```python
from pyspark.sql.types import IntegerType
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from pyspark.sql.functions import split, explode, udf, collect_list, broadcast, col, max
from pyspark.sql.types import ArrayType, StringType, IntegerType
import random
from hashlib import sha256
from Levenshtein import distance
from collections import defaultdict, Counter
import os
import shutil
```

```python
spark = SparkSession.builder \
    .appName("Process Trace Analysis") \
    .getOrCreate()

df = spark.read.csv('/content/drive/My Drive/log.txt', header=False, inferSchema=True).toDF('FromServer', 'ToServer', 'Time', 'Action', 'CaseId')

df.show()
```

# visalization of the process for each caseId

```python
# showing traces of each caseId
windowSpec = Window.partitionBy('CaseId').orderBy('Time')

df_traces = df.withColumn("Trace", F.collect_list("ToServer").over(windowSpec))
df_traces = df_traces.groupby("CaseId").agg(F.concat_ws(", ", F.collect_list("ToServer")).alias("Trace"))

df_traces.show(truncate=False)
```

# k-Shingles

```python
# Define a function to generate 2-shingles
def generate_shingles(trace):
    return [f"{{{trace[i]}, {trace[i+1]}}}" for i in range(len(trace)-1)]

# Register the function as a UDF
udf_generate_shingles = F.udf(generate_shingles, ArrayType(StringType()))

# Split the 'Trace' column into a list
df_shingles = df_traces.withColumn("Trace", F.split(df_traces["Trace"], ", "))

df_shingles = df_shingles.withColumn("Shingles", udf_generate_shingles("Trace"))

df_shingles.select("CaseId", "Shingles").show(truncate=False)
```

```python
# Flatten the 'Shingles' column into a new DataFrame
df_flat = df_shingles.select(F.explode(df_shingles.Shingles).alias("Shingles"))

# Get all unique 2-shingles
df_unique_shingles = df_flat.distinct()

# Collect all unique 2-shingles into a list
unique_shingles_list = df_unique_shingles.rdd.flatMap(lambda x: x).collect()

print(unique_shingles_list)
```

```python
shingle_dict = {shingle: idx for idx, shingle in enumerate(unique_shingles_list)}
shingle_dict_broadcast = spark.sparkContext.broadcast(shingle_dict)

def shingles_to_vector(shingles):
    vector = [0] * len(shingle_dict_broadcast.value)
    for shingle in shingles:
        if shingle in shingle_dict_broadcast.value:
            vector[shingle_dict_broadcast.value[shingle]] = 1
    return vector

vector_udf = udf(shingles_to_vector, ArrayType(IntegerType()))
df_binary = df_shingles.withColumn("BinaryVector", vector_udf("Shingles"))
df_binary.select("CaseId", "BinaryVector").show(truncate=False)
```

```python
# Step 1: Define a UDF to simulate hash functions (permutations)
def minhash_signature(vector, num_hashes, seed=123):
    random.seed(seed)
    length = len(vector)
    signatures = []
    for _ in range(num_hashes):
        # Create a random permutation of indices
        permutation = list(range(length))
        random.shuffle(permutation)
        # Find the index of the first '1' in the permuted vector
        min_index = next((i for i in permutation if vector[i] == 1), None)
        signatures.append(min_index)
    return signatures

minhash_udf = udf(minhash_signature, ArrayType(IntegerType()))


num_hashes = 20

df_signatures = df_binary.withColumn("MinHashSignatures", minhash_udf("BinaryVector", F.lit(num_hashes)))

df_signatures.select("CaseId", "MinHashSignatures").show(truncate=False)
df_signatures.select("MinHashSignatures").distinct().show(truncate=False)
```

# LSH

```python
num_hashes = 20  # Assuming this from your MinHash
b = 4           # Number of bands
r = 5           # Rows per band (should satisfy b * r = num_hashes)

def hash_band(band):
    band_string = '-'.join(map(str, band))  # string format
    return sha256(band_string.encode('utf-8')).hexdigest()

hash_band_udf = F.udf(hash_band, StringType())

for i in range(b):
    df_signatures = df_signatures.withColumn(f'BandHash_{i+1}', hash_band_udf(F.slice(df_signatures["MinHashSignatures"], i * r + 1, r)))

df_signatures.show(truncate=False)

all_candidate_pairs = []
for i in range(b):
    band_col = f'BandHash_{i+1}'
    candidates = (df_signatures.groupBy(band_col)
                             .agg(collect_list("CaseId").alias("Candidates"))
                             .filter(F.size("Candidates") > 1))
    print(f"Candidates from band {i+1}:")
    candidates.show(truncate=False)
    # Now, for each band, generate all pairs of candidates within each list
    for row in candidates.collect():
        candidate_list = row["Candidates"]
        for i in range(len(candidate_list)):
            for j in range(i+1, len(candidate_list)):
                all_candidate_pairs.append((candidate_list[i], candidate_list[j]))

print(all_candidate_pairs)
```

```python
# Reduce the number of DataFrame accesses by collecting traces once
trace_dict = {row['CaseId']: row['Trace'] for row in df_signatures.select("CaseId", "Trace").collect()}

groups = defaultdict(list)

def find_group(candidate, groups):
    for group, members in groups.items():
        if candidate in members:
            return group
    return None

threshold = 20 # Set the similarity threshold for Edit Distance

# Compare each pair's Edit Distance and group them if similar
for pair in all_candidate_pairs:
    trace1 = trace_dict[pair[0]]
    trace2 = trace_dict[pair[1]]
    edit_distance = distance(trace1, trace2)

    if edit_distance < threshold:
        group1 = find_group(pair[0], groups)
        group2 = find_group(pair[1], groups)

        if group1 and group2:
            if group1 != group2:
                groups[group1].extend(groups.pop(group2))
        elif group1:
            groups[group1].append(pair[1])
        elif group2:
            groups[group2].append(pair[0])
        else:
            groups[f"Group_{len(groups)}"].extend([pair[0], pair[1]])

# Print the groups of similar items
for group, processes in groups.items():
    print(f"{group}: {processes}")
```

```python
# Convert groups dictionary to DataFrame
group_data = [(group, processes) for group, processes in groups.items()]
df_groups = spark.createDataFrame(group_data, ["Group_id", "Containing_Cases"])

# Explode Containing_Cases to join with df_traces
df_group_expanded = df_groups.withColumn("CaseId", explode(col("Containing_Cases")))

# Join with df_traces to get traces for each CaseId in groups
df_group_with_traces = df_group_expanded.join(df_traces, "CaseId", "inner")

# Define a UDF to find the most common CaseId
def most_common_case(cases):
    case_counter = Counter(cases)  # Count the occurrences of each CaseId
    most_common = case_counter.most_common(1)[0][0]  # Get the most common CaseId
    return most_common

most_common_case_udf = udf(most_common_case, IntegerType())

df_final = df_group_with_traces.groupBy("Group_id").agg(
    most_common_case_udf(collect_list("CaseId")).alias("Representative_Case"),
    collect_list("CaseId").alias("Containing_Cases")
)

df_final.show(truncate=False)
```

```python
representative_cases = set(df_final.select("Representative_Case").rdd.flatMap(lambda x: x).collect())

# Filter df to keep only the rows where 'CaseId' is in representative_cases
df_filtered = df.filter(col("CaseId").isin(representative_cases)).select("FromServer", "ToServer", "Time", "Action", "CaseId")

# Write the DataFrame to a CSV file in a temporary directory
df_filtered.coalesce(1).write.csv("part1Output_temp", mode='overwrite')

part_file = [filename for filename in os.listdir('part1Output_temp') if filename.startswith('part-00000')][0]
full_path = os.path.join('part1Output_temp', part_file)

final_output_path = 'Final1Output.txt'
shutil.move(full_path, final_output_path)
```

```python
with open('part1Observations.txt', 'w') as file:
    for row in df_final.collect():
        file.write(f"Group: {row['Containing_Cases']}\n")

        for case in row['Containing_Cases']:
            file.write(f"{case}:\n")

            logs = df.filter(df['CaseId'] == case).collect()

            for log in logs:
                file.write(f"< {log['FromServer']}, {log['ToServer']}, {log['Time']}, {log['Action']}, {log['CaseId']} >\n")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import StringIndexer, OneHotEncoder
from pyspark.ml.evaluation import ClusteringEvaluator
import matplotlib.pyplot as plt

spark = SparkSession.builder.appName("KMeans Clustering").getOrCreate()
data = spark.read.csv("Final1Output.txt", inferSchema=True, header=True)
data = data.toDF("FromServer", "ToServer", "Time", "Action", "CaseId")


# Convert the 'Action' column to an index
indexer = StringIndexer(inputCol="Action", outputCol="ActionIndex")
indexed = indexer.fit(data).transform(data)

# One-hot encode the 'ActionIndex' column
encoder = OneHotEncoder(inputCols=["ActionIndex"], outputCols=["ActionVec"])
model = encoder.fit(indexed)
encoded = model.transform(indexed)

vec_assembler = VectorAssembler(inputCols=["FromServer", "ToServer", "Time", "CaseId"], outputCol="features")
final_data = vec_assembler.transform(data)


scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
scaler_model = scaler.fit(final_data)
cluster_final_data = scaler_model.transform(final_data)

kmeans = KMeans(featuresCol='scaledFeatures',k=2)
model = kmeans.fit(cluster_final_data)

predictions = model.transform(cluster_final_data)
evaluator = ClusteringEvaluator()
silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))

centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)

predictions = model.transform(cluster_final_data)
predictions.select("prediction").show()
```

```python
silhouettes = []
ks = list(range(2, 10))  # Replace with the range of k you want to try

for k in ks:
    kmeans = KMeans(featuresCol='scaledFeatures', k=k)
    model = kmeans.fit(cluster_final_data)
    predictions = model.transform(cluster_final_data)
    silhouette = evaluator.evaluate(predictions)
    silhouettes.append(silhouette)

plt.plot(ks, silhouettes)
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.show()
```
